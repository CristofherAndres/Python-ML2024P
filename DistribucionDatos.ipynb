### Distribuciones de datos
- Distribución uniforme

```python
import numpy as np
#Importar librería para gráficos
import matplotlib.pyplot as plt

datos = np.random.uniform(0, 5, 250)
plt.hist(datos, 5)
plt.show()
```

- Distribución normal

```python
import numpy as np
#Importar librería para gráficos
import matplotlib.pyplot as plt

datos = np.random.normal(5, 1, 250)
plt.hist(datos, 100)
plt.show()
```
