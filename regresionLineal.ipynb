```python
import matplotlib.pyplot as plt
from scipy import stats

x = [50, 18, 25, 32, 55, 20, 19]
y = [5, 0, 2, 4, 4, 1, 1]



slope, intercept, r, p, se = stats.linregress(x, y)

def rectaPrediccion(x):
    return slope * x + intercept


print("La relacion es: ", r)
print("El error del modelo es: ", se)

miModelo = list(map(rectaPrediccion, x))

plt.scatter(x,y)
plt.plot(x, miModelo, c='r')
plt.show()
```

```python
#Alternativa con datos random

import matplotlib.pyplot as plt
from scipy import stats
import numpy as np


x = np.random.normal(50,10,1000)
y = np.random.normal(5,2, 1000)




slope, intercept, r, p, se = stats.linregress(x, y)

def rectaPrediccion(x):
    return slope * x + intercept

print("La relacion es: ", r)
print("El error del modelo es: ", se)


miModelo = list(map(rectaPrediccion, x))

plt.scatter(x,y)
plt.plot(x, miModelo, c='r')
plt.show()
```
