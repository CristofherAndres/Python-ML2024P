```python
import matplotlib.pyplot as plt

edades = [18, 22, 30, 22,44,34,54,23,65,46,56,34]
autos = [0,2,1,0,5,6,3,1,0,2,1,0]

plt.scatter(edades,autos, label='Edades vs Autos', color='r', s=25, marker="x")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

edad = np.random.uniform(18,65,1000)
autos = np.random.uniform(0,10, 1000)

plt.scatter(edad,autos, color='g', s=25)
plt.show()


edad = np.random.normal(50,10,1000)
autos = np.random.normal(5,2, 1000)

plt.scatter(edad,autos, color='r', s=25)
plt.show()
```
