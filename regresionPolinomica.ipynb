### Regresi√≥n Polinomica

```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]


miModelo = np.poly1d(np.polyfit(x,y,3))

miLinea = np.linspace(1,22, 50)

print(sklearn.metrics.r2_score(y,miModelo(x)))

plt.scatter(x,y, color='g',label='X vs Y')
plt.plot(miLinea, miModelo(miLinea))
plt.legend()
plt.show()
```
